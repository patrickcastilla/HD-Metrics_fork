For the logistic distribution $\Lambda(x)=(1+\exp (-x))^{-1}$ verify that

 (a) [1 points] $\frac{d}{d x} \Lambda(x)=\Lambda(x)(1-\Lambda(x))$.

**_Answer_ :**

Sea $(1-\Lambda(x)) =  1-\left(1+\exp(-x)\right)^{-1}\ =(\exp(-x))\left(1 + \exp(-x)\right)^{-1}$. Entonces, 

$$
\begin{aligned}
& \frac{\partial \Lambda(x)}{\partial x}=(-1)\left(1+\exp(-x)\right)^{-2}\left(-\exp(-x)\right) \\
& \quad \quad \quad =\exp(-x)\left(1 + \exp(-x)\right)^{-2} \\
& \quad \quad \quad = \left(1 + \exp(-x)\right)^{-1}(\exp(-x))\left(1 + \exp(-x)\right)^{-1} \\
& \quad \quad \quad = \Lambda(x)(1-\Lambda(x))
\end{aligned}
$$

(b) [1 points] $h_{\operatorname{logit}}(x)=\frac{d}{d x} \log \Lambda(x)=1-\Lambda(x)$.

**_Answer_ :**

Sea $\log \Lambda(x) = -1 \log(1+\exp(-x))$

$$
\begin{aligned}
& h_{\operatorname{logit}}(x)=\frac{d}{d x} \log \Lambda(x)\\
& \quad \quad \quad = (-1)(1 + \exp(-x))^{-1}(-\exp(-x)) \\
& \quad \quad \quad = (\exp(-x))(1+\exp(-x) )^{-1}\\
& \quad \quad \quad = (1-\Lambda(x))
\end{aligned}
$$

(c) [1 points] $H_{\text {logit }}(x)=-\frac{d^{2}}{d x^{2}} \log \Lambda(x)=\Lambda(x)(1-\Lambda(x))$.

**_Answer_ :**
$$
\begin{aligned}
& H_{\text {logit }}(x)=-\frac{d^{2}}{d x^{2}} \log \Lambda(x) \\
& \quad \quad \quad = (-1)[((-\exp(-x))(1 + \exp(-x))-(\exp(x))](1+\exp(-x))^{-2} \\
& \quad \quad \quad = (-1)[(-\exp(-x))(1+\exp(-x) -\exp (-x))(1+\exp(-x) )^{-2}\\
& \quad \quad \quad = (-1) (-\exp(-x))(1+\exp(-x))^{-2}\\
& \quad \quad \quad = (\exp(-x))(1+\exp(-x))^{-2}\\
& \quad \quad \quad = \Lambda(x) (1-\Lambda(x))
\end{aligned}
$$

(d) [1 points] $\left|H_{\text {logit }}(x)\right| \leq 1$.

**_Answer_ :**

Sabemos que $ 0\leq \Lambda(x) \leq 1$
$$
\begin{aligned}
    0 \leq \Lambda(x) \leq 1 \\
     0 \leq 1-\Lambda(x) \leq 1 \\
     0 \leq \Lambda(x)(1-\Lambda(x)) \leq 1 \\

     \left|H_{\text {logit }}(x)\right| \leq 1
\end{aligned}
$$

#### Question 3

[5 points] The observations are $\{\left(Y_i, X_{1 i}, X_{2 i}\right)\}, i=1, \ldots, n$. You estimate two least squares regressions.

$$
\begin{gathered}
Y_i=X_{1 i}^{\prime} \tilde{\beta}_1+\tilde{e}_i \text{,}\\
Y_i=X_{1 i}^{\prime} \hat{\beta}_1+X_{2 i}^{\prime} \hat{\beta}_2+\hat{e}_i^{\prime}\text{,}
\end{gathered}
$$

and calculate the residual variance estimates

$$
\begin{aligned}
\tilde{\sigma}^2 & =\frac{1}{n} \sum_{i=1}^n \tilde{e}_i^2\text{,} \\
\hat{\sigma}^2 & =\frac{1}{n} \sum_{i=1}^n \hat{e}_i^2\text{.}
\end{aligned}
$$


Show that for any $w \in(0,1)$, there is a constant $a \in(0,1)$ such that

$$
\frac{1}{n} \sum_{i=1}^n\left(w \hat{e}_i+(1-w) \tilde{e}_i\right)^2=(1-a) \hat{\sigma}^2+a \tilde{\sigma}^2\text{.}
$$


Find the constant $a$.

**_Answer_ :**

Desarollando la expresión:

$$n^{-1} \sum_{i=1}^n\left(w \hat{e}_i+(1-w) \tilde{e}_i\right)^2 = w^2 n^{-1} \sum_{i=1}^n \hat{e}_i^2+2 w(1-w) n^{-1} \sum_{i=1}^n \hat{e}_i \tilde{e}_i+(1-w)^2 n^{-1} \sum_{i=1}^n\tilde{e}_i^2$$
$$n^{-1} \sum_{i=1}^n\left(w \hat{e}_i+(1-w) \tilde{e}_i\right)^2= w^2 \hat{\sigma}^2+2 w(1-w) n^{-1} \sum_{i=1}^n \hat{e}_i \tilde{e}_i+(1-w)^2 \tilde{\sigma}^2$$
    
Tenemos que:
    $$
    n^{-1} \sum_{i=1}^n \hat{e}_i \tilde{e}_i=n^{-1} \hat{e}^{\prime} \tilde{e}
    $$
    
La matriz generadora de residuos de $\hat{e}_i$ es $\boldsymbol{M}=\boldsymbol{I}_n-\boldsymbol{X}\left(\boldsymbol{X}^{\prime} \boldsymbol{X}\right)^{-1} \boldsymbol{X}$, donde $\boldsymbol{X}=\left[\boldsymbol{X}_1 \boldsymbol{X}_2\right]$.
    
La matriz generadora de residuos de $\tilde{e}_i$ es $\boldsymbol{M}_1=\boldsymbol{I}_n-\boldsymbol{X}_1\left(\boldsymbol{X}_1 \boldsymbol{X}_1\right)^{-1} \boldsymbol{X}_1$.
    
$$n^{-1} \hat{e}^{\prime} \tilde{e}=n^{-1} \boldsymbol{Y}^{\prime} \boldsymbol{M}^{\prime} \boldsymbol{M}_1 \boldsymbol{Y}=n^{-1} \boldsymbol{Y}^{\prime} \boldsymbol{M} \boldsymbol{M}_1 \boldsymbol{Y}$$
$$\boldsymbol{M} \boldsymbol{M}_1=(\boldsymbol{I}-\boldsymbol{P})\left(\boldsymbol{I}-\boldsymbol{P}_1\right)=\left(\boldsymbol{I}-\boldsymbol{P}-\boldsymbol{P}_1+\boldsymbol{P} \boldsymbol{P}_1\right)$$
    
Dado que $\boldsymbol{X}_1$ es parte de $\boldsymbol{X}$, $\boldsymbol{P} \boldsymbol{P}_1=\boldsymbol{P}_1$
$$
\boldsymbol{M} \boldsymbol{M}_1=\left(\boldsymbol{I}-\boldsymbol{P}-\boldsymbol{P}_1+\boldsymbol{P}_1\right)=(\boldsymbol{I}-\boldsymbol{P})=\boldsymbol{M}
$$
    
Luego:
$$
\frac{1}{n} \boldsymbol{Y}^{\prime} \boldsymbol{M} \boldsymbol{M}_1 \boldsymbol{Y}=\frac{1}{n} \boldsymbol{Y}^{\prime} \boldsymbol{M} \boldsymbol{Y}=\frac{1}{n} \boldsymbol{Y}^{\prime} \boldsymbol{M} \boldsymbol{M} \boldsymbol{Y}=\frac{1}{n} \hat{e}^{\prime} \hat{e}=\hat{\sigma}^2
$$
    
Entonces:
$$
\begin{aligned}
& (1-a) \hat{\sigma}^2+a \hat{\sigma}^2=w^2 \hat{\sigma}^2+2 w(1-w) \hat{\sigma}^2+(1-w)^2 \tilde{\sigma}^2 \\
& (1-a) \hat{\sigma}^2+a \tilde{\sigma}^2=\left(w^2+2 w(1-w)\right) \hat{\sigma}^2+(1-w)^2 \tilde{\sigma}^2
\end{aligned}
$$
    
Notar que $(w+(1-w))^2=w^2+2 w(1-w)+(1-w)^2=1$. Luego:
$$(1-a) \hat{\sigma}^2+a \tilde{\sigma}^2=\left(1-(1-w)^2\right) \hat{\sigma}^2+(1-w)^2 \tilde{\sigma}^2$$
$$a=(1-w)^2$$

$$
\tilde{\beta}=\left(\sum_{i=1}^n X_i X_i^{\prime} \mathbb{1}\left\{\|X_i\| \leq c\right\}\right)^{-1}\left(\sum_{i=1}^n X_i Y_i \mathbb{1}\left\{\|X_i\| \leq c\right\}\right)

(a) [2 points] Show that $\tilde{\beta} \underset{p}{\longrightarrow} \beta$.


**_Answer_ :**

$$\tilde{\beta}=\left(\sum_{i=1}^{n}X_{i}X_{i}^{\prime}\mathbb{1}\left\{ \left|X_{i}\right|\leq c\right\} \right)^{-1}\left(\sum_{i=1}^{n}X_{i}Y_{i}\mathbb{1}\left\{\left|X_{i}\right|\leq c\right\} \right)$$

$$\mathbb{E}[\tilde{\beta}\mid\mathbf{X}]=\beta$$
$$\tilde{\beta}\underset{p}{\rightarrow}\beta$$

**_Answer_ :**

$$\hat{\beta} = \left( \frac{1}{n} \sum_{i=1}^n X_i X_i^\prime \mathbb{1}\{|X_i| \leq c\} \right)^{-1} \left( \frac{1}{n} \sum_{i=1}^n X_i Y_i \mathbb{1}\{|X_i| \leq c\} \right)$$

$$\hat{\beta}=\left(\boldsymbol{X}^{\prime} \boldsymbol{X}+\lambda \boldsymbol{I}_{p}\right)^{-1}\left(\boldsymbol{X}^{\prime} \boldsymbol{Y}\right)=\hat{\beta}_{\text {ridge}}$$
